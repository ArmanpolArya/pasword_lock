```python
password = "1975"  
tries = 3           

while tries > 0:
    enter = input("Please enter password: ")
    if enter == password:
        print("Unlock")
        break
    else:
        tries -= 1
        if tries > 0:
            print("Incorrect. You have", tries, "tries left.")
        else:
            print("Incorrect. No tries left. Locked out.")
```
